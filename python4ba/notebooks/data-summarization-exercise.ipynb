## Data frame with columns

- year,month,day
        Date of departure    
- dep_time,arr_time
        Actual departure and arrival times (format HHMM or HMM), local tz.
- sched_dep_time,sched_arr_time
        Scheduled departure and arrival times (format HHMM or HMM), local tz.    
- dep_delay,arr_delay
        Departure and arrival delays, in minutes. Negative times represent early departures/arrivals.
- hour,minute
        Time of scheduled departure broken into hour and minutes.
- carrier
        Two letter carrier abbreviation. See airlines() to get name
- tailnum
        Plane tail number
- flight
        Flight number
- origin,dest
        Origin and destination. See airports() for additional metadata.
- air_time
        Amount of time spent in the air, in minutes
- distance
        Distance between airports, in miles
- time_hour
        Scheduled date and hour of the flight as a date. Along with origin, can be used to join flights data to weather data.

```python
# Find all flights (identfied by 'carrier' and 'flight') 
# that always depart at least 60 minutes late.
df=flights.groupby(['carrier','flight']).agg({'dep_delay':'min'})
df[df.dep_delay>=60]
```

```python
# How many flights always depart at least 60 minutes late? 53
df=flights.groupby(['carrier','flight']).agg({'dep_delay':'min'})
df[df.dep_delay>=60].count()
```

```python
# Finda all United Airlines (UA) flights (identfied by 'carrier' and 'flight') 8
# that depart (at least) 60 minutes late.
#flights[(flights.carrier=='UA')&(flights.dep_delay > 60)].groupby(['carrier','flight']).size()
df=flights[(flights.carrier=='UA')].groupby(['carrier','flight']).agg({'dep_delay':'min'})

df[df.dep_delay>=60]
```

```python
# How many United Airlines (UA) flights depart (at least) 60 minutes late? 3824
flights[(flights.dep_delay>60)& (flights.carrier == 'UA')] 
```

```python
# Flights that always arrived on time in December? 
df=flights[(flights.month==12)].groupby(['carrier','flight']).agg({'arr_delay':'min'})
df[df.arr_delay<0]
```

```python
# How many flights always arrived on time in December? 1855
flights[(flights.arr_delay<0)&(flights.month==12)].groupby(['carrier','flight']).size()
```

```python
# Create a dataframe 'cancelled' for all cancelled flights. 
# Assumption: flights that are not cancelled should have values for dep_time. 
can = pd.DataFrame(flights[flights.dep_time.isnull()])
can
```

```python
# Look at the number of cancelled flights per day. 
# Find the top 10 days with the most cancelled flights. 
(
    can.groupby(['year','month','day'])
    .agg({'flight':'count'})
    .sort_values('flight',ascending=False)
    .head(10)
)
```

```python
# Which carrier has the worst delays? 
flights.groupby('carrier')['dep_delay'].max().sort_values(ascending=False)
flights.groupby('carrier')['arr_delay'].max().sort_values(ascending=False)
```

```python
# Which plane (tailnum) has the worst on-time record? N844MH
flights[flights.arr_delay.notnull()].groupby('tailnum')['arr_delay'].mean().sort_values(ascending=False)
```

```python
# What time (hour) of day should you fly if you want to avoid delays as much as possible?
flights.groupby('hour')['dep_delay'].mean().sort_values()
flights.groupby('hour')['arr_delay'].mean().sort_values()
```

```python
# For each destination, compute the total minutes of arrival delay. 
flights.groupby('dest')['arr_delay'].sum()
```
